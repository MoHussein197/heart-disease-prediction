```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# ===============================
# 1. Load Dataset
# ===============================

# Path to dataset (place heart_disease.csv in data/ folder)
data_path = '../data/heart_disease.csv'
df = pd.read_csv(data_path)

# Display first 5 rows
print(df.head())

# ===============================
# 2. Data Exploration
# ===============================
print("\nDataset shape:", df.shape)
print("\nDataset info:")
print(df.info())
print("\nSummary statistics:")
print(df.describe())

# Target distribution
if 'num' in df.columns:
    print("\nTarget distribution:")
    print(df['num'].value_counts())

# ===============================
# 3. Data Cleaning
# ===============================

# Replace '?' with NaN and convert columns to numeric
for col in ['ca','thal']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop rows with missing values
df = df.dropna()

# ===============================
# 4. Encode categorical features
# ===============================

# Convert binary columns to int
binary_cols = ['sex','fbs','exang']
for col in binary_cols:
    df[col] = df[col].astype(int)

# One-hot encode multi-category columns
df = pd.get_dummies(df, columns=['cp','restecg','slope','thal'], drop_first=True)

# ===============================
# 5. Scaling
# ===============================

# MinMax Scaler
minmax_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(minmax_scaler.fit_transform(df), columns=df.columns)

# Standard Scaler
std_scaler = StandardScaler()
df_standard = pd.DataFrame(std_scaler.fit_transform(df), columns=df.columns)

# ===============================
# 6. Save Preprocessed Data
# ===============================

df.to_csv('../data/heart_disease_clean.csv', index=False)
print("\n✅ Preprocessed data saved as heart_disease_clean.csv")
```
