```python
# ===============================
# 6. Unsupervised Learning
# ===============================

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

df = pd.read_csv("../data/heart_disease_clean.csv")
X = df.drop("target", axis=1)
y = df["target"]

X_scaled = StandardScaler().fit_transform(X)

# KMeans Elbow
inertia = []
for k in range(1, 10):
    km = KMeans(n_clusters=k, random_state=42).fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(range(1,10), inertia, marker="o")
plt.title("Elbow Method - KMeans")
plt.show()

# Compare clusters
kmeans = KMeans(n_clusters=2, random_state=42).fit(X_scaled)
print("KMeans vs Actual:\n", pd.crosstab(kmeans.labels_, y))

# Hierarchical clustering
Z = linkage(X_scaled, method="ward")
dendrogram(Z)
plt.title("Hierarchical Dendrogram")
plt.show()

agg = AgglomerativeClustering(n_clusters=2, linkage="ward").fit(X_scaled)
print("Agglomerative vs Actual:\n", pd.crosstab(agg.labels_, y))
```
